```python
from collections import defaultdict

# Crie um dicionário para armazenar a contagem de cada idioma
contagem_idiomas = defaultdict(int)

# Itere pelas linhas do DataFrame
for idiomas in df['languages']:
  # Se a linha não for nula
  if not pd.isnull(idiomas):
    # Divida a string de idiomas em uma lista de idiomas individuais
    lista_idiomas = idiomas.split(',')
    # Para cada idioma na lista
    for idioma in lista_idiomas:
      # Remova espaços em branco e atualize a contagem no dicionário
      idioma = idioma.strip()
      contagem_idiomas[idioma] += 1

# Converta o dicionário em uma série do pandas e ordene por frequência
idiomas_mais_falados = pd.Series(contagem_idiomas).sort_values(ascending=False)

# Exiba o resultado (top 10, por exemplo)
print(idiomas_mais_falados.head(10))
```
